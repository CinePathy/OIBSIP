# STEP 5: VISUALIZATION

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Calories'].mean().plot(kind='bar')
plt.ylabel("Average Calories")
plt.title("Average Calories per Food Category")
plt.show()
```

```python
plt.boxplot(df['Calories'])
plt.title("Calories Distribution")
plt.show()
```

```python
nutrition = df[['Calories', 'Total Fat', 'Sodium', 'Carbohydrates', 'Protein']]
sns.heatmap(nutrition.corr(), annot=True)
plt.show()
```
