# STEP 1: DATA PREPARATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apps = pd.read_csv(
    r"C:\Users\lokes\OneDrive\Desktop\DA 11-google play store\apps.csv"
)
```

```python
reviews = pd.read_csv(
    r"C:\Users\lokes\OneDrive\Desktop\DA 11-google play store\user.csv"
)
```

```python
apps.columns = apps.columns.str.lower().str.replace(" ", "_")
#fix column names
```

```python
apps.dropna(inplace=True)
#handling missing values
```

# STEP 2: Category Exploration

```python
category_count = apps['category'].value_counts()
#no. of aapps per category
```

```python
category_count.plot(kind='bar', figsize=(10,5))
plt.title("App Distribution Across Categories")
plt.show()
```

```python
#Certain categories like Family, Games and Tools dominate the Play Store.
```

#  STEP 3: Metrics Analysis

```python
apps['rating'].mean()
```

```python
#install vs ratings
sns.scatterplot(x='rating', y='installs', data=apps)
plt.show()
```

```python
#paid vs free apps
apps.groupby('type')['rating'].mean()
```

# STEP 4: Sentiment and Pricing Analysis

```python
apps_paid = apps[apps['type'] == 'Paid']
apps_paid['price'].describe()
```

```python
sns.boxplot(x=apps_paid['price'])
plt.show()
```

```python
# Most paid apps are priced low, indicating price sensitivity.
```

```python
reviews['Sentiment'].value_counts()
```

# Step 5: Interactive visualization

```python
#Category vs rating
sns.barplot(x='category', y='rating', data=apps)
plt.xticks(rotation=90)
plt.title("Average Rating by App Category")
plt.show()
```

```python
#install vs rating
sns.scatterplot(x='rating', y='installs', data=apps)
plt.title("Installs vs Ratings")
plt.show()
```
